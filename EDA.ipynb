```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv("amz_uk_price_prediction_dataset.csv")
```

```python
#Generar una tabla de frecuencias para el producto category.
```

```python
df.sample(10)
```

```python
df.category.value_counts()
```

```python
freq_category = df.category.value_counts()
```

```python
freq_category.head(5)
```

```python
top5 = freq_category.head()
```

```python
top5
```

#### Sports & Outdoors, Beauty, Handmade Clothing, Shoes & Accessories, Bath & Body, Birthday Gifts

```python
plt.figure(figsize=(10, 5))
sns.barplot(x=top5.index, y=top5.values)
plt.title("Distribución de Productos en las Principales Categorías")
plt.xlabel("Categoría de Producto")
plt.ylabel("Frecuencia")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.pie(top5, labels=top5.index, autopct="%1.1f%%")
plt.title("Proporción de Productos en las Principales Categorías")
plt.show()
```

```python
mean_price = df["price"].mean().round()
mean_price
```

```python
median_price = df["price"].median()
median_price 
```

```python

mode_price = df["price"].mode()[0]
mode_price 
```

```python

price_std_dev = round( df["price"].std(),2)
price_std_dev
```

```python

price_variance = round(df["price"].var(),2)
price_variance
```

```python

price_range = df["price"].max() - df["price"].min()
price_range
```

```python

iqr_price = df["price"].quantile(0.75) - df["price"].quantile(0.25)
iqr_price
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(df["price"], bins=30, kde=True)
plt.title("Distribución de Precios de Productos")
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df["price"])
plt.title("Box Plot de Precios de Productos")
plt.xlabel("Precio")
plt.show()
```

```python

median_stars = df["stars"].median()
median_stars
```
